# <center>El problema de kepler<center>
## <center>(Problema de dos cuerpos para una atraccion central)<center>
    
    
### Una breve introduccion que nos permite conocer el tipo de dinamicas que las masas pueden describir 

Consiste en encontrar la posición de dos cuerpos (Uno de masa mucho mayor que el segundo) a lo largo del tiempo implementando únicamente sus masas y posiciones iniciales.

<div class="alert alert-block alert-warning">
<b>Ejemplo:</b> Un ejemplo del problema de Kepler es describir el movimiento de la tierra alrededor del sol.
</div>

<img src=https://astrojem.com/imagenes_voltaire/tierraejerotacion.gif>

Algunas de las suposiciones necesarias para resolver el problema de Kepler son:
<ol>
<li>Las masas deben tener una contextura totalmente esférica.</li>
<li>Una de las masas debe tener una masa mucho mas grande que la otra.</li>
<li>No existe algún tipo de fricción en el movimiento de los cuerpos.</li>
</ol>

Por comodidad, las ecuaciones referentes a los movimientos de dos o más objetos se miden a partir de la distancia $r$ que hay entre ellos, por ende, las coordenadas polares permiten realizar un estudio mas sencillo de las posiciones que establecen a través del tiempo. 

La energía $E$ es la suma de la energía cinetica $T$ y la energía potencial $U$

$$E=T+U$$

Donde $T$ es el trabajo es proporcional a la de masa $m$ del cuerpo multiplicado por el vector velocidad $\dot{\vec{r}}$ al cuadrado.

$$T=\frac{1}{2}m (\dot{\vec{r}})^2$$

y la energía $U$ es inversamente proporcional a la distancia entre las masas 

$$U=\frac{k}{r}$$

<div class="alert alert-block alert-info">
<b>Nota:</b> La derivada del vector posición $\vec{r}$ o velocidad, es igual a $\dot{\vec{r}}=\dot{r}^2+r^2\dot{\theta}^2$.
</div>

Por ende la energia se expresa de la forma:
    
$$E=\frac{1}{2}m (\dot{r}^2+r^2\dot{\theta}^2)+\frac{k}{r}$$

# Momento ángular 

Al momento angular $L$ lo definiremos como el producto entre la masa del objeto, la distancia que hay entre los objetos al cuadrado y el cambio del Angulo a través del tiempo.

$$L=m r^2 \dot{\theta}$$

Por lo tanto el cambio del Angulo a través del tiempo es igual a: 
$$\dot{\theta}=\frac{L}{mr^2}$$

# Planteamiento de la ecuación diferencial

De las expresiones de momento angular y energía se obtiene que:

$$E=\frac{1}{2}m \frac{dr}{dt}^2+\frac{L^2}{2mr^2}+\frac{k}{r}$$

Despejando $\frac{dr}{dt}$ en lo anterior

$$\frac{dr}{dt}=\sqrt{\frac{2}{m}\left(E-\frac{L^2}{2mr^2}-\frac{k}{r}\right)}$$

Al dividir la derivada de la distancia entre la derivada del angulo con respecto al tiempo se tiene la siguiente ecuación diferencial en términos de $r$ y $\theta$  
$$\frac{d\theta}{d r}=\frac{\frac{L}{mr^2}}{\sqrt{\frac{2}{m}\left(E-\frac{L^2}{2mr^2}-\frac{k}{r}\right)}}$$


# Solución real al problema de kepler 

La solución real del problema de kepler es igual a 

$$r=\frac{-\frac{L^2}{km}}{1-\sqrt{1+\frac{2EL^2}{k^2m}}cos(\theta)}$$ 

Donde la excentricidad $e$ es de la forma 

$$e=\sqrt{1+\frac{2EL^2}{k^2m}}$$

y $p$ el parametro focal 

$$p=-\frac{L^2}{km}$$

La solución de la ecuacion diferencial $r$ describe una seccion conica en coordenadas polares, la cual se identifica gracias al valor de $p$ y $e$.

$$r=\frac{p}{1-e cos(\theta)}$$ 


Determinar que la solución de las ecuaciones de kepler son cónicas nos permite dar completa seguridad de que las trayectorias de las masas de distribución mas pequeña pueden ser de trayectoria hiperbólica, sin embargo, se requiere identificar bajo qué condiciones podemos obtener el diferente tipo de cónicas, lo cual se establecerá mas adelante en un análisis posterior

## Métodos numéricos para resolver ecuaciones diferenciales

En la sección anterior se pudo apreciar como a partir de principios físicos sencillos se podía describir la dinámica de una masa orbitando alrededor de una masa inconmensurablemente mayor, una ecuación que daba solución al problema clásico de Kepler solucionada analíticamente, Sin embargo, para 3 o más cuerpos, las dinámicas descritas por los cuerpos se modelan a partir de una cantidad mayor de variables, las cuales se modelan empleando ecuaciones vectoriales cuya solución requiere de métodos numéricos los cuales describiremos a continuación: 

### Runge kutta de orden 1

Runge kutta es un método numérico iterativo utilizado para la solución de ecuaciones diferenciales de orden 1. Su planteamiento es el siguiente

$$y_{k+1}=y_k+hf(x_k,y_k)$$

<div class="alert alert-block alert-warning">
<b>Ejemplo:</b> Runge kutta de orden 1 aplicado a una ecuacion aleatoria para comprobar su funcionamiento. Se usara la ecuación  
    
$$\frac{dy}{dx}=x^2-3y$$
    
con las condiciones iniciales $$y(0)=1$$


Cuya solucion real es $$y(x)=\frac{25}{27}e^{-3x}+\frac{1}{3}\left(x^2-\frac{2}{3}x+\frac{2}{3}\right)$$
</div>


```python
def f(x,y):#Se declara la funcion para el metodo de runge kutta
    dydx=x**2-3*y
    return dydx
```

```python
n=500#numero de pasos en la funcion solución
h=4/n#tamaño de paso

x0=0#x inicial
y0=1#y inicial 

solx=np.array([x0])#declarar vector solucion de x  
soly=np.array([y0])#declarar vector solucion de y


#evaluar en la solucion real
x=np.linspace(0,4,500)
y=25/27*np.exp(-3*x)+1/3*(x**2-2/3*x+2/9)
```

```python
for i in range(n):
    y0=y0+h*f(x0,y0)#expresion de runge kutta
    x0=x0+h#aumentar el x a partir del tamaño de paso
    solx=np.append(solx,x0)#agregar el valor en el vector solucion
    soly=np.append(soly,y0)#agregar el valor en el vector solucion
```

```python
#graficar solución
fig, ax = plt.subplots(1,2, figsize=(18, 8))
ax[0].plot(solx,soly)
ax[0].set_title('Solución por el método de runge kutta')
#graficar solución analitica
ax[1].plot(x,y)
ax[1].set_title("Solución analitica")
plt.show()
```

### Runge kutta de cuarto orden

$$y_{k+1}=y_k+\frac{h}{6}(K_1+2K_2+2K_3+K_4)$$

$$K_1=f\left(x_k,y_k\right)$$

$$K_2=f\left(x_k+\frac{h}{2},y_k+\frac{h}{2}K_1\right)$$

$$K_3=f\left(x_k+\frac{h}{2},y_k+\frac{h}{2}K_2\right)$$

$$K_1=f\left(x_k+h,y_k+hK_3\right)$$

<div class="alert alert-block alert-warning">
<b>Ejemplo:</b> Runge kutta de orden 4 aplicado a una ecuacion aleatoria para comprobar su funcionamiento. Se usara la ecuación  
    
$$\frac{dy}{dx}=x^2-3y$$
    
con las condiciones iniciales $$y(0)=1$$


Cuya solucion real es $$y(x)=\frac{25}{27}e^{-3x}+\frac{1}{3}\left(x^2-\frac{2}{3}x+\frac{2}{3}\right)$$
</div>

```python
n=500#cantidad de puntos de runge kutta
h=4/n#tamaño de paso

x0=0#x inicial 
y0=1#y inicial 

solx=np.array([x0])#definir e inicializar vectores iniciales
soly=np.array([y0])

for i in range(n):
    k1=f(x0,y0)#calcular k1
    k2=f(x0+h/2,y0+h*k1/2)#calcular k2
    k3=f(x0+h/2,y0+h*k2/2)#calcular k3
    k4=f(x0+h,y0+h*k3)#calcular k4
    y0=y0+h*(k1+2*k2+2*k3+k4)/6#calcular valor aproximado de y
    x0=x0+h#calcular valor del proximo x
    solx=np.append(solx,x0)#agregar los valores en el vector
    soly=np.append(soly,y0)    
```

```python
#graficar solución
fig, ax = plt.subplots(1,2, figsize=(18, 8))
ax[0].plot(solx,soly)
ax[0].set_title("Solucion por el metodo de runge kutta de cuarto orden")
#graficar solución analitica
ax[1].plot(x,y)
ax[1].set_title("Solución analitica")
plt.show()
```

## Modelo diferencial que describe las dinámicas de n cuerpos

Las dinámicas de movimiento planetario tienen como base principal la segunda ley de newton, la cual establece que la aceleración que experimenta un cuerpo es proporcional a las fuerzas que recibe. $$F=m\times a$$

Para este caso, la aceleración esta descrita por la segunda derivada con respecto al tiempo de la distancia entre los cuerpos $i$ y $j$, denotada por $\frac{d^2 \overrightarrow{r_{i.j}}}{dt^2}$, por lo tanto, la fuerza descrita se obtiene de multiplicar una de las masas del sistema $m_i$ por la aceleración

$$\overrightarrow{F}=m_i\times\frac{d^2 \overrightarrow{r_i}}{dt^2}$$

Newton estableció que la magnitud de la fuerza de atracción entre dos cuerpos perfectamente esféricos es proporcional al producto de sus masas e inversamente proporcional al cuadrado de la magnitud de sus distancias, lo cual se conoce como ley de gravitación universal, definida matemáticamente como:

$$F=G\frac{m_1m_2}{|r_{1,2}|^2}$$

Donde $G$ es la constante de gravitación universal, determinada por Henry Cavendish en 1798. 

Para obtener el componente vectorial de la fuerza de atracción, se multiplica la magnitud de esta por un vector unitario de la distancia de las masas, es decir

$$\overrightarrow{F}=G\frac{m_1m_2}{|r_{1,2}|^2}\frac{\overrightarrow{r_{1,2}}}{|r_{1,2}|}=G\frac{m_1m_2}{|r_{1,2}|^3}\overrightarrow{r_{1,2}}$$

Para $n$ cuerpos, la magnitud de fuerza obtenida se obtiene como la suma de todas las fuerzas implicadas, es decir:

$$\overrightarrow{F}=G\sum_{i\neq j}^n \frac{m_im_j}{|r_{i,j}|^3}\overrightarrow{r_{i,j}}$$

Donde $\overrightarrow{r_{i,j}}$ indica la distancia que hay entre la masa $i$ y la masa $j$

Finalmente, al relacionar las dos expresiones de fuerza planteadas anteriormente se obtiene:

$$m_i\frac{d^2 \overrightarrow{r_i}}{dt^2}=G\sum_{i\neq j}^n \frac{m_im_j}{|r_{i,j}|^3}\overrightarrow{r_{i,j}}$$


### Solución numérica para dos cuerpos

La ley de gravitacion universal no solo proporciona la posición de un cuerpo orbitando, sino tambien la velocidad del movimiento, para ello, se debe resolver la ecuacion diferencial usando un sistema de ecuaciones equivalente, de esta manera, considere el siguiente cambio de variables de polares a cartesianas:

$$|r_{1,2}|=\sqrt{x^2+y^2}$$

$$\overrightarrow{r_{1,2}}=(x\overrightarrow{i}+y\overrightarrow{j})$$
Cuyo resultado es

$$\frac{d^2(x\overrightarrow{i}+y\overrightarrow{j})}{dt^2}=G\frac{m_1}{(\sqrt{x^2+y^2})^3}(x\overrightarrow{i}+y\overrightarrow{j})$$

La cual puede ser separada en dos ecuaciones únicamente dependientes de $x$ e $y$ respectivamente 

$$\frac{d^2x}{dt^2}=\frac{Gm_1x}{(\sqrt{x^2+y^2})^3}$$

$$\frac{d^2y}{dt^2}=\frac{Gm_1y}{\sqrt{(x^2+y^2})^3}$$

Para simplificar este sistema, se separará en 4 expresiones diferentes denotando nuevas variables $v_x,v_y$ como las derivadas de $x$ e $y$

$$\frac{dx}{dt}=v_x$$

$$\frac{dy}{dt}=v_y$$

$$\frac{dv_x}{dt}=\frac{Gm_1x}{(\sqrt{x^2+y^2})^3}$$

$$\frac{dv_y}{dt}=\frac{Gm_1y}{(\sqrt{x^2+y^2})^3}$$


```python
#t es la medida del tiempo 
#Y es un vector que contiene a y=(x,y,dx,dy)
#f es la funcion que representa a la ecuacion diferencial
def f(t,Y):
    xprime = Y[2]
    yprime = Y[3]
    vxprime = Y[0]*G*m1/np.power(Y[0]**2+Y[1]**2,3/2)
    vyprime = Y[1]*G*m1/np.power(Y[0]**2+Y[1]**2,3/2)
    return np.array([xprime,yprime,vxprime,vyprime])

#rungeKuttaTwo resuelve la ecuación diferencial f para un tiempoTotal definido, 
#una cantidad total de puntos n
#una masa m1
#un tiempo inicial t
#condiciones iniciales Y
def rungeKuttaTwo(tiempoTotal,n,m1,t,Y,f):
    h=tiempoTotal/n #tamaño de paso
    solTiempo=np.zeros(n)#declarar un vector de tiempo compuesto por ceros 
    solY=np.zeros((n,4)) #declarar un vector solución Y compuesto por n vectores de 4 elementos 
    
    for i in range(n):
        solTiempo[i]=t
        solY[i]=Y
        k0=f(t,Y)
        k1=f(t+h/2,Y+h*k0/2)
        k2=f(t+h/2,Y+h*k1/2)
        k3=f(t+h,Y+h*k2)
        DY=h/6*(k0+2*k1+2*k2+k3)
    
        t=t+h #incremento en t
        Y=Y+DY #incremento en el vector Y
        #print(Y,DY)
    return solY[:,0],solY[:,1],solY[:,2],solY[:,3],solTiempo
```

```python
tiempoTotal=1#tiempo total
n=1000 #numero de puntos
m1=10 #masa 10
t=0 #tiempo inicial
G=6.674-11#Constante de gravitación universal
Y=np.array([1,1,1,-3])#condiciones iniciales
x,y,vx,vy,time=rungeKuttaTwo(tiempoTotal,n,m1,t,Y,f)
```

Para un cuerpo de masa 10 en un tiempo inicial $0$ posición inicial $(1,1)$ velocidad inicial horizontal $1$ y velocidad inicial vertical $-3$ posee una dinámica elíptica alrededor del sol posicionado en el punto $(0,0)$ durante un tiempo de $1$ unidad temporal.

La velocidad horizontal y vertical describen un comportamiento periódico que oscila entre valores positivos y negativos que dan sentido al comportamiento elíptico del planeta

```python
def f3D(t,Y):
    xprime = Y[3]
    yprime = Y[4]
    zprime = Y[5]
    vxprime = Y[0]*G*m1/np.power(Y[0]**2+Y[1]**2+Y[2]**2,3/2)
    vyprime = Y[1]*G*m1/np.power(Y[0]**2+Y[1]**2+Y[2]**2,3/2)
    vzprime = Y[2]*G*m1/np.power(Y[0]**2+Y[1]**2+Y[2]**2,3/2)
    return np.array([xprime,yprime,zprime,vxprime,vyprime,vzprime])

#rungeKuttaTwo resuelve la ecuación diferencial f para un tiempoTotal definido, 
#una cantidad total de puntos n
#una masa m1
#un tiempo inicial t
#condiciones iniciales Y
def rungeKutta3D(tiempoTotal,n,m1,t,Y,f):
    h=tiempoTotal/n #tamaño de paso
    solTiempo=np.zeros(n)#declarar un vector de tiempo compuesto por ceros 
    solY=np.zeros((n,6)) #declarar un vector solución Y compuesto por n vectores de 4 elementos 
    
    for i in range(n):
        solTiempo[i]=t
        solY[i]=Y
        k0=f(t,Y)
        k1=f(t+h/2,Y+h*k0/2)
        k2=f(t+h/2,Y+h*k1/2)
        k3=f(t+h,Y+h*k2)
        DY=h/6*(k0+2*k1+2*k2+k3)
    
        t=t+h #incremento en t
        Y=Y+DY #incremento en el vector Y
        #print(Y,DY)
    return solY[:,0],solY[:,1],solY[:,2],solY[:,3],solY[:,4],solY[:,5],solTiempo
```

# Tres cuerpos

Obtener las dinámicas de los cuerpos en movimiento junto con sus velocidades se resolvió aplicando un sistema simple de Kepler, suponiendo una masa muy pequeña orbitando alrededor del sol independiente del cuerpo menor hiperbólico que aproxima el planeta, debido al efecto de atracción despreciable que este provoca en el resto de los cuerpos. Sin embargo, se estudia la posibilidad de aplicar una generalización a la solución de Kepler para el problema de dos cuerpos. El problema de tres cuerpos, con el cual se ha escrito un código especializado en generar las trayectorias de dichos cuerpos.
En la imagen generada a continuación, se podrá ver como tres masas de masas similares modifican su trayectoria inicial y se acercan. La trayectoria de color rojo representando una masa de $10$ unidades de posición inicial (2,2) y velocidad inicial (1,-3) curva su dirección a la determinada por los otros dos cuerpos, mientras que la trayectoria amarilla con masa $10$ de posición inicial (2,1) y velocidad inicial (1,-3) se aproxima con dirección al objeto azul de masa $5$ posición inicial (1,2) y velocidad (1,-3) modifica su trayectoria por completo al verse atraída por la fuerza gravitacional de la masa amarilla y roja, dirigiéndose a una segura colisión con el cuerpo de trayectoria amarilla 


Si bien es interesante el plantear un elemento adicional al sistema, el proceso de análisis tiene una mayor complejidad, por lo que se podría considerar agregarlo una vez este completo el análisis de dos cuerpos y ver en que aspectos coincidan los sistemas 


```python
def f(t,Y):
    xprime1 = Y[6]
    yprime1 = Y[7]
    xprime2 = Y[8]
    yprime2 = Y[9]
    xprime3 = Y[10]
    yprime3 = Y[11]
    
    vxprime1 = m2*G*(Y[0]-Y[2])/np.power((Y[0]-Y[2])**2+(Y[1]-Y[3])**2,3/2)+m3*G*(Y[0]-Y[4])/np.power((Y[0]-Y[4])**2+(Y[1]-Y[5])**2,3/2)
    vyprime1 = m2*G*(Y[1]-Y[3])/np.power((Y[0]-Y[2])**2+(Y[1]-Y[3])**2,3/2)+m3*G*(Y[1]-Y[5])/np.power((Y[0]-Y[4])**2+(Y[1]-Y[5])**2,3/2)
    
    vxprime2 = m1*G*(Y[2]-Y[0])/np.power((Y[2]-Y[0])**2+(Y[3]-Y[1])**2,3/2)+m3*G*(Y[2]-Y[4])/np.power((Y[2]-Y[4])**2+(Y[3]-Y[5])**2,3/2)
    vyprime2 = m1*G*(Y[3]-Y[1])/np.power((Y[2]-Y[0])**2+(Y[3]-Y[1])**2,3/2)+m3*G*(Y[3]-Y[5])/np.power((Y[2]-Y[4])**2+(Y[3]-Y[5])**2,3/2)
    
    vxprime3 = m1*G*(Y[4]-Y[0])/np.power((Y[4]-Y[0])**2+(Y[5]-Y[1])**2,3/2)+m3*G*(Y[4]-Y[2])/np.power((Y[4]-Y[2])**2+(Y[5]-Y[3])**2,3/2)
    vyprime3 = m1*G*(Y[5]-Y[1])/np.power((Y[4]-Y[0])**2+(Y[5]-Y[1])**2,3/2)+m3*G*(Y[5]-Y[3])/np.power((Y[4]-Y[2])**2+(Y[5]-Y[3])**2,3/2)
    return np.array([xprime1,yprime1,xprime2,yprime2,xprime3,yprime3,vxprime1,vyprime1,vxprime2,vyprime2,vxprime3,vyprime3])
```

```python
def rungeKuttaKepler(tiempoTotal,n,m1,m2,m3,t,Y,f):
    h=tiempoTotal/n #tamaño de paso
    solTiempo=np.zeros(n)#declarar un vector de tiempo compuesto por ceros 
    solY=np.zeros((n,np.size(Y))) #declarar un vector solución Y compuesto por n vectores de 4 elementos 
    
    for i in range(n):
        solTiempo[i]=t
        solY[i]=Y
        k0=f(t,Y)
        k1=f(t+h/2,Y+h*k0/2)
        k2=f(t+h/2,Y+h*k1/2)
        k3=f(t+h,Y+h*k2)
        DY=h/6*(k0+2*k1+2*k2+k3)
    
        t=t+h #incremento en t
        Y=Y+DY #incremento en el vector Y
        #print(Y,DY)
    return solY,solTiempo
```

```python
tiempoTotal=0.14#tiempo total
n=1000 #numero de puntos
m1=5 #masa 10
m2=10
m3=10
t=0 #tiempo inicial
G=6.674-11#Constante de gravitación universal
#Y=np.array([1,2,3,3,4,1,1,-1,0,-1,-1,1])#condiciones iniciales

Y=np.array([2,2,2,1,1,2,
            1,-3,1,-3,1,-3])#condiciones iniciales
Sol,time=rungeKuttaKepler(tiempoTotal,n,m1,m2,m3,t,Y,f)
```
